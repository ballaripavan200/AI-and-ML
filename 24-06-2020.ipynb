```python
# preparing the input and output columns
x=boston_df[['RM']]# selecting columns a dataframe
y=boston_df[['MEDV']]
x
```

```python
#plt.style.available
#plt.style.use
with plt.style.context("grayscale"):
    plt.plot(x,y,"bo")
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("scatter plot")
```

```python
## usual sklearn import,instantiate ,fit
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x,y)
```

```python
## print intercept and coefficient and coefficient 
print(lm.intercept_)
print(lm.coef_)
```

# Interpreting the model:
- A unit increase in X(RM)  is associated with a 9.10210898 unit increase in MEDV

```python
## with help of predict , we are going to find our predictedd values 
y_pred=lm.predict(x)
y_pred
```

```python
## r2_value-->to check the model
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
## observe the regresion line with sns.axes_style('dark'):
sns.regplot(x,y)
```
