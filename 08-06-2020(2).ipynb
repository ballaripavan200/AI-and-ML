### Merging (inner,outer,left,right)

```python
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'], "total":[90,96,98,92]})
df1
```

```python
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'], "distance_from_city":[40,60,12,20]})
df2
```

```python
pd.merge(df1,df2,on="branch",how='inner')
```

```python
pd.merge(df1,df2,on="branch",how='outer')
```

```python
pd.merge(df1,df2,on="branch",how='left')
```

```python
pd.merge(df1,df2,on="branch",how='right')
```

```python
df1
```

```python
df2
```

```python
df3=pd.merge(df1,df2,on="branch",how='inner')
```

```python
pd.concat([df1,df3],axis=1)
```
